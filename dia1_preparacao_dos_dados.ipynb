```python
import pandas as pd

url = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_1-Importando_dados/Datasets/dados_emprestimos/emprestimos-20111.csv'
df = pd.read_csv(url)
df.head()
```

```python
url_acervo = 'https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet'
acervo_df = pd.read_parquet(url_acervo)
acervo_df.head()
```

```python
df.duplicated().sum()
acervo_df.duplicated().sum()
```

```python
df['codigo_barras'].isnull().sum()
acervo_df['codigo_barras'].isnull().sum()
```

```python
df_unificado = pd.merge(df, acervo_df,
                        on='codigo_barras', how= 'inner')
```
